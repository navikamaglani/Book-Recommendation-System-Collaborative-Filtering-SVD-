## Collaborative Filtering Based Recommender System

```python
## Collaborative Filtering Based Recommender System

# Filtering users who have rated more than 200 books
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users = x[x].index
```

```python
# Filtering ratings of the users who have rated more than 200 books
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
# Filtering books with at least 50 ratings
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
# Final filtered ratings DataFrame with only famous books
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
# Creating a pivot table with Book-Title as rows and User-ID as columns
pt = final_ratings.pivot_table(index= 'Book-Title', columns='User-ID',values='Book-Rating')
pt
```

```python
# Filling NaN values with 0 in the pivot table
pt.fillna(0, inplace=True)
```

```python
# Importing cosine_similarity from sklearn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calculating the cosine similarity between books
cosine_similarity(pt).shape
```

```python
# Calculating the similarity scores
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
# Defining the recommend function to get book recommendations based on similarity
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

### Calculating RMSE for the Recommendation Model using Cosine Similarity

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
train_data, test_data = train_test_split(final_ratings, test_size=0.2, random_state=42)
```

```python
# Create a pivot table for the training set
train_pivot = train_data.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
train_pivot.fillna(0, inplace=True)
```

```python
# Calculate similarity scores using cosine similarity
similarity_scores_cosine = cosine_similarity(train_pivot)
```

```python
def predict_ratings(user_id, book_title, similarity_scores, train_pivot):
    if book_title not in train_pivot.index:
        return 0  
    if user_id not in train_pivot.columns:
        return 0 
    
    # Find the index of the book
    book_index = train_pivot.index.get_loc(book_title)
    
    # Calculate the weighted sum of ratings
    similarity_scores_book = similarity_scores[book_index]
    user_ratings = train_pivot.loc[:, user_id]
    
    # Handle case where user has no ratings
    if user_ratings.sum() == 0:
        return user_ratings.mean()
    
    weighted_sum = sum(similarity_scores_book * user_ratings)
    sim_sum = sum(similarity_scores_book)
    
    # Handle division by zero
    if sim_sum == 0:
        return user_ratings.mean()
    
    predicted_rating = weighted_sum / sim_sum
    return predicted_rating
```

```python
from sklearn.metrics import mean_squared_error

# Predict ratings for the test set using cosine similarity
test_data['predicted_rating'] = test_data.apply(
    lambda x: predict_ratings(x['User-ID'], x['Book-Title'], similarity_scores_cosine, train_pivot), 
    axis=1
)

# Calculate RMSE for cosine similarity
rmse_cosine = np.sqrt(mean_squared_error(test_data['Book-Rating'], test_data['predicted_rating']))
print(f'RMSE using Cosine Similarity: {rmse_cosine}')
```

### Calculating RMSE for the Recommendation Model using Euclidean Distance

```python
from sklearn.metrics.pairwise import pairwise_distances

# Calculate similarity scores using Euclidean distance
similarity_scores_euclidean = 1 / (1 + pairwise_distances(train_pivot, metric='euclidean'))
```

```python
# Predict ratings for the test set using Euclidean distance
test_data['predicted_rating_euclidean'] = test_data.apply(lambda x: predict_ratings(x['User-ID'], x['Book-Title'], similarity_scores_euclidean, train_pivot), axis=1)
```

```python
# Calculate RMSE
rmse_euclidean = np.sqrt(mean_squared_error(test_data['Book-Rating'], test_data['predicted_rating_euclidean']))
print(f'RMSE using Euclidean Distance: {rmse_euclidean}')
```

### Results Comparison

- **RMSE using Cosine Similarity**: 3.898
- **RMSE using Euclidean Distance**: 3.952

### Interpretation

The lower RMSE indicates better performance in predicting the ratings. In this case:

- The RMSE for **Cosine Similarity** is slightly lower than for Euclidean Distance, suggesting that the Cosine Similarity metric provides more accurate predictions for this recommendation system.

### Conclusion

Based on the RMSE values, the recommendation model using **Cosine Similarity** performs better compared to using **Euclidean Distance** for predicting book ratings in this dataset.

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity


# Adjust filtering criteria
# Filter users who have rated at least 10 books
user_counts = ratings['User-ID'].value_counts()
active_users = user_counts[user_counts >= 10].index
filtered_ratings = ratings[ratings['User-ID'].isin(active_users)]

# Filter books that have at least 20 ratings
book_counts = filtered_ratings['ISBN'].value_counts()
popular_books = book_counts[book_counts >= 20].index
filtered_ratings = filtered_ratings[filtered_ratings['ISBN'].isin(popular_books)]

# Add the specific book back if it was filtered out
isbn_to_check = '034545104X'
if isbn_to_check not in filtered_ratings['ISBN'].unique():
    specific_book_ratings = ratings[ratings['ISBN'] == isbn_to_check]
    filtered_ratings = pd.concat([filtered_ratings, specific_book_ratings])

# Create user-item matrix
user_item_matrix = filtered_ratings.pivot(index='User-ID', columns='ISBN', values='Book-Rating').fillna(0)
item_user_matrix = user_item_matrix.T

# Calculate cosine similarity
item_similarity = cosine_similarity(item_user_matrix)

# Convert to DataFrame for easy handling
item_similarity_df = pd.DataFrame(item_similarity, index=item_user_matrix.index, columns=item_user_matrix.index)

# Function to get top-N similar books
def get_similar_books(book_isbn, top_n=5):
    if book_isbn not in item_similarity_df.columns:
        print(f"Book ISBN {book_isbn} not found in the dataset.")
        return []
    similar_books = item_similarity_df[book_isbn].sort_values(ascending=False).head(top_n + 1).index.tolist()
    similar_books.remove(book_isbn)  # Remove the book itself
    return similar_books

# Example: Get top 5 books similar to a specific book ISBN
similar_books = get_similar_books(isbn_to_check, top_n=5)
print(f"Books similar to {isbn_to_check}:", similar_books)
```

### Recommendation

```python
def recommend_books(book_title, similarity_df, pt, n=5):
    # Get the index of the book that matches the title
    book_idx = pt.index.get_loc(book_title)
    
    # Get the similarity scores for this book with all other books
    similarity_scores = similarity_df.iloc[book_idx]
    
    # Sort the books by their similarity scores in descending order
    sorted_indices = similarity_scores.argsort()[::-1]
    
    # Get the top n similar books
    top_n_books = pt.index[sorted_indices[1:n+1]]
    return top_n_books

#Get top 5 book recommendations for '1984'
recommendations_without_dim_reduction = recommend_books('1984', similarity_df, pt, n=5)
recommendations_without_dim_reduction
```

## Ratings Without Dimensionality Reduction:

```python
def predict_ratings(similarity_df, pt):
    mean_user_rating = pt.mean(axis=1).values  # Convert to numpy array
    ratings_diff = (pt.T - mean_user_rating).T
    similarity_sum = np.array([np.abs(similarity_df).sum(axis=1)]).T
    pred = mean_user_rating[:, np.newaxis] + (similarity_df.dot(ratings_diff.T).T / similarity_sum.T)
    return pred

pred_without_dim_reduction = predict_ratings(similarity_df, pt)
```

## Ratings With Dimensionality Reduction:

```python
pred_with_dim_reduction = predict_ratings(reduced_similarity_df, pt)
```

```python
def calculate_rmse(predictions, actuals):
    # Convert DataFrames to NumPy arrays
    predictions = predictions.to_numpy()
    actuals = actuals.to_numpy()
    
    # Get the non-zero indices
    non_zero_indices = actuals.nonzero()
    
    # Extract the predicted and actual values at these indices
    pred_values = predictions[non_zero_indices].flatten()
    actual_values = actuals[non_zero_indices].flatten()
    
    # Calculate the RMSE
    return np.sqrt(((pred_values - actual_values) ** 2).mean())

rmse_without_dim_reduction = calculate_rmse(pd.DataFrame(pred_without_dim_reduction), pt)
rmse_with_dim_reduction = calculate_rmse(pd.DataFrame(pred_with_dim_reduction), pt)
```
